```python
import numpy as np
```

```python
#downloading the file to jupyter notebook
#converting a csv file into ndarray
taxi = np.genfromtxt('nyc_taxis.csv', delimiter=',',skip_header=True)
```

```python
taxi
```

```python
#mean speed of taxis
speed = taxi[:, 7]/ (taxi[:, 8]/3600) #7th column has trip distance and 8th column has trip time in sec
mean_speed = np.mean(speed)
mean_speed
```

```python
#number of rides taken in february
rides_feb = taxi[taxi[:, 1] == 2, 1] #gives only one column the rides taken in month february
print(rides_feb.shape[0])
```

```python
#no. of rides where tip more than $50
print(taxi[taxi[:, -3] > 50, -3].shape[0])
```

```python
#no. of rides where drop was JFK airport
#dropoff location code for jfk airport is 2
print(taxi[taxi[:, 6] == 2, 6].shape[0])
```
